```python
from welly import Well
from welly import Curve
import welly.quality as wq
from IPython.display import HTML

import pandas as pd
import matplotlib.pyplot as plt
```

```python
well = Well.from_las('49025104000000_283139.LAS')
```

```python
tests = {'Each': [wq.no_flat, wq.no_monotonic, wq.no_gaps]}
```

```python
data_qc_table = well.qc_table_html(tests)
HTML(data_qc_table)
```

```python
tests_2 = {'Each': [wq.no_flat, wq.no_monotonic, wq.no_gaps], 
           'GRD': [wq.all_positive], 
           'RHOB': [wq.all_positive, 
                    wq.all_between(1.5, 3),
                    wq.check_units(['G/CC', 'g/cm3']),
        ]}
```

```python
data_qc_table = well.qc_table_html(tests_2)
HTML(data_qc_table)
```

```python
tests_nans = {'Each': [wq.fraction_not_nans]}

data_nans_qc_table = well.qc_table_html(tests_nans)
HTML(data_nans_qc_table)
```

```python
print((f'Curve \t % Complete').expandtabs(10))
print((f'----- \t ----------').expandtabs(10))

for k,v in well.qc_data(tests_nans).items():
    
    for i,j in v.items():
        values = round(j*100, 2)
    print((f'{k} \t {values}%').expandtabs(10))
```

```python
well.data['GRD_DESPIKE'] = well.data['GRD'].despike(window_length=50, z=2)
```

```python
well.data['GRD_CLIP'] = well.data['GRD'].clip(0, 120)
```

```python
well.plot(tracks=['MD', ['GRD', 'GRD_CLIP']])
```

```python
well.data['GRD_FILTER'] = well.data['GRD']._rolling_window(window_length = 50, step =2)
```
